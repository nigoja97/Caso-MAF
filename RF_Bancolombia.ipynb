```python
# Crear una lista de tickers
tickers = ["BIL", "BILS", "IEF", "IEI", "SHY", "IGSB", "IGIB", "IGLB", "HYG", "CEMB", "FLOT", "MBB", "EBND", "JPMB", "VWOB", "EMB", "VTIP", "STIP", "TIP"]

# Obtener datos históricos para cada ticker
data = yf.download(tickers, period="1y")["Adj Close"]

# Calcular los retornos logarítmicos mensuales
log_returns_monthly = np.log(data / data.shift(21))  # Usamos 21 días hábiles como aproximación a un mes

# Calcular la matriz de covarianza
covariance_matrix = log_returns_monthly.cov()

# Número de activos
n_assets = len(tickers)

# Definir los pesos como variables de optimización
weights = cp.Variable(n_assets)

# Definir el objetivo de la función de optimización (minimizar la varianza)
portfolio_variance = cp.quad_form(weights, covariance_matrix)

# Restricciones: los pesos deben sumar 1 y ser no negativos
constraints = [cp.sum(weights) == 1, weights >= 0]

# Definir el problema de optimización y resolverlo
try:
    problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
    problem.solve()

    # Verificar el estado del problema de optimización
    if problem.status == cp.OPTIMAL:
        # Obtener los pesos óptimos asignados a cada activo
        optimal_weights = weights.value

        # Calcular el rendimiento esperado para cada activo
        expected_returns = log_returns_monthly.mean().values

        # Calcular el rendimiento esperado del portafolio óptimo
        portfolio_expected_return = np.sum(expected_returns * optimal_weights)

        # Calcular los puntos de la frontera eficiente
        n_points = 100
        target_returns = np.linspace(min(expected_returns), max(expected_returns), n_points)
        efficient_frontier_variances = []
        for target_return in target_returns:
            target_return = np.array([target_return])
            constraints = [
                cp.sum(weights) == 1,
                weights >= 0,
                expected_returns @ weights == target_return
            ]
            problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
            problem.solve()
            if problem.status != cp.OPTIMAL:
                raise ValueError("El problema de optimización no se resolvió de manera óptima.")
            efficient_frontier_variances.append(np.sqrt(portfolio_variance.value))

        # Convertir los valores de volatilidad a porcentaje con dos cifras decimales
        efficient_frontier_variances = np.round(np.array(efficient_frontier_variances) * 100, 2)

        # Convertir los valores de rendimiento esperado a porcentaje con dos cifras decimales
        target_returns = np.round(np.array(target_returns) * 100, 2)

    else:
        print("El problema de optimización no se resolvió de manera óptima.")
except Exception as e:
    print("Error al resolver el problema de optimización:", e)

# Obtener el índice del punto con el coeficiente de Sharpe máximo
max_sharpe_index = np.argmax(np.array(target_returns) / np.array(efficient_frontier_variances))

# Obtener el rendimiento esperado y la volatilidad en el punto con el coeficiente de Sharpe máximo
max_sharpe_return = target_returns[max_sharpe_index]
max_sharpe_volatility = efficient_frontier_variances[max_sharpe_index]

# Obtener los pesos de los activos en el punto con el coeficiente de Sharpe máximo
weights_max_sharpe = optimal_weights

# Graficar la frontera eficiente
plt.figure(figsize=(10, 6))
plt.scatter(efficient_frontier_variances, target_returns, marker='o', color='blue', label='Frontera Eficiente')
plt.scatter(max_sharpe_volatility, max_sharpe_return, marker='*', color='red', s=200, label='Máxima Relación Rendimiento/Volatilidad')
plt.title('Frontera Eficiente (Retornos mensuales)')
plt.xlabel('Volatilidad (%)')
plt.ylabel('Rendimiento Esperado (%)')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir el rendimiento esperado y la volatilidad en el punto con el coeficiente de Sharpe máximo
print("Rendimiento Esperado en el Punto Máximo de Sharpe:", max_sharpe_return)
print("Volatilidad en el Punto Máximo de Sharpe:", max_sharpe_volatility)

# Imprimir los pesos de los activos en el punto con el coeficiente de Sharpe máximo
print("Pesos de los Activos en el Punto Máximo de Sharpe:")
for ticker, weight in zip(tickers, weights_max_sharpe):
    print(f"Ticker: {ticker}, Peso: {weight:.4f}")
```

```python
# Crear una lista de tickers
tickers = ["BIL", "BILS", "IEF", "IEI", "SHY", "IGSB", "IGIB", "IGLB", "HYG", "CEMB", "FLOT", "MBB", "EBND", "JPMB", "VWOB", "EMB", "VTIP", "STIP", "TIP"]

# Obtener datos históricos para cada ticker
data = yf.download(tickers, period="3y")["Adj Close"]

# Calcular los retornos logarítmicos anuales
log_returns_annual = np.log(data / data.shift(252))  # Usamos 252 días hábiles como aproximación a un año

# Calcular la matriz de covarianza
covariance_matrix = log_returns_annual.cov()

# Número de activos
n_assets = len(tickers)

# Definir los pesos como variables de optimización
weights = cp.Variable(n_assets)

# Definir el objetivo de la función de optimización (minimizar la varianza)
portfolio_variance = cp.quad_form(weights, covariance_matrix)

# Restricciones: los pesos deben sumar 1 y ser no negativos
constraints = [cp.sum(weights) == 1, weights >= 0]

# Definir el problema de optimización y resolverlo
try:
    problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
    problem.solve()

    # Verificar el estado del problema de optimización
    if problem.status == cp.OPTIMAL:
        # Obtener los pesos óptimos asignados a cada activo
        optimal_weights = weights.value

        # Calcular el rendimiento esperado para cada activo
        expected_returns = log_returns_annual.mean().values

        # Calcular el rendimiento esperado del portafolio óptimo
        portfolio_expected_return = np.sum(expected_returns * optimal_weights)

        # Calcular los puntos de la frontera eficiente
        n_points = 100
        target_returns = np.linspace(min(expected_returns), max(expected_returns), n_points)
        efficient_frontier_variances = []
        for target_return in target_returns:
            target_return = np.array([target_return])
            constraints = [
                cp.sum(weights) == 1,
                weights >= 0,
                expected_returns @ weights == target_return
            ]
            problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
            problem.solve()
            if problem.status != cp.OPTIMAL:
                raise ValueError("El problema de optimización no se resolvió de manera óptima.")
            efficient_frontier_variances.append(np.sqrt(portfolio_variance.value))

        # Convertir los valores de volatilidad a porcentaje con dos cifras decimales
        efficient_frontier_variances = np.round(np.array(efficient_frontier_variances) * 100, 2)

        # Convertir los valores de rendimiento esperado a porcentaje con dos cifras decimales
        target_returns = np.round(np.array(target_returns) * 100, 2)


    else:
        print("El problema de optimización no se resolvió de manera óptima.")
except Exception as e:
    print("Error al resolver el problema de optimización:", e)


# Obtener el índice del punto con el coeficiente de Sharpe máximo
max_sharpe_index = np.argmax(np.array(target_returns) / np.array(efficient_frontier_variances))

# Obtener el rendimiento esperado y la volatilidad en el punto con el coeficiente de Sharpe máximo
max_sharpe_return = target_returns[max_sharpe_index]
max_sharpe_volatility = efficient_frontier_variances[max_sharpe_index]

# Obtener los pesos de los activos en el punto con el coeficiente de Sharpe máximo
weights_max_sharpe1 = optimal_weights

# Graficar la frontera eficiente
plt.figure(figsize=(10, 6))
plt.scatter(efficient_frontier_variances, target_returns, marker='o', color='green', label='Frontera Eficiente')
plt.scatter(max_sharpe_volatility, max_sharpe_return, marker='*', color='red', s=200, label='Máxima Relación Rendimiento/Volatilidad')
plt.title('Frontera Eficiente (Retornos Anuales)')
plt.xlabel('Volatilidad (%)')
plt.ylabel('Rendimiento Esperado (%)')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir el rendimiento esperado y la volatilidad en el punto con el coeficiente de Sharpe máximo
print("Rendimiento Esperado en el Punto Máximo de Sharpe:", max_sharpe_return)
print("Volatilidad en el Punto Máximo de Sharpe:", max_sharpe_volatility)

# Imprimir los pesos de los activos en el punto con el coeficiente de Sharpe máximo
print("Pesos de los Activos en el Punto Máximo de Sharpe:")
for ticker, weight in zip(tickers, weights_max_sharpe1):
    print(f"Ticker: {ticker}, Peso: {weight:.4f}")

```
